```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import mlxtend
from mlxtend.frequent_patterns import association_rules, apriori, fpgrowth
```

```python
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_excel("Selldata.xlsx",sheet_name = "Data set")
```

```python
data.head()
```

```python
df = data
```

```python
df = df.drop(columns = "Foods")
```

```python
dfsumfood = pd.DataFrame()
```

```python
food_columns = df.columns[df.columns.str.contains('food', case=False)]
```

```python
food_columns
```

```python
dffood = df[food_columns]
```

```python
dffood.head()
```

```python
#Cleansing data
```

```python
dffood = dffood.astype(str)
```

```python
dfsumfood['Items'] = dffood[food_columns].apply(lambda x: ','.join(x), axis=1)
```

```python
dfsumfood
```

```python
basket = dfsumfood['Items'].str.get_dummies(",")
```

```python
basket.columns
```

```python
basket.head()
```

```python
basket.isnull().sum()
```

```python
basket = basket.astype(pd.SparseDtype(bool))
```

```python
basket.head(10)
```

```python
#Drop columns
```

```python
header = ["nan"," "," 20"," 2016 "," 2017","Water","Focaccia Salsa complimentary","Hamper D 2 Panettone","Olive","Rasberry Sherbet REVIEW",
          "Special food","Fried dried chilli.galic.seafood",
         "Hamper Special set","Lemon Sherbet review",
          "Stir Fried White Wine with Drili Seafood","Special drink","Set B 1890","Pasta + Set",
         "Set D 2790","Special Set 599","French Fries Side Dish","Mash Potato Side Dish","Hamper A Happiness",
          "Foccacia Herbs Salsa 3 pcs","Set C 2190","Mixed Oilves","Special Set 999"]
```

```python
basket = basket.drop(columns = header)
```

```python
basket.head(10)
```

```python
basket.describe()
```

```python
basket.columns
```

```python
basket.sum()
```

```python
#Association Rules
```

```python
itemlist = fpgrowth(basket, min_support = 0.01, use_colnames = True)
itemlist.sort_values('support', ascending=False)
```

```python
metric='support'
rules = association_rules(itemlist, metric = metric, min_threshold = 0.01)
rules.sort_values('support', ascending = False, inplace = True)
rules
```

```python
rules['antecedents'] = rules['antecedents'].apply(lambda x: ', '.join(list(x)))
rules['consequents'] = rules['consequents'].apply(lambda x: ', '.join(list(x)))
```

```python
rules
```

```python
with pd.ExcelWriter('italyrestaurant_1year_results.xlsx') as writer:
    rules.to_excel(writer, sheet_name='Sheet1', index=False)
```
